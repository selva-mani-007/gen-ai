```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
```

```python
# Load the Stable Diffusion model from Hugging Face
model_id = "runwayml/stable-diffusion-v1-5"  # Public model

# Load the model pipeline
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")  # Move model to GPU if available
```

```python
# Provide a text prompt for image generation
prompt = "peacock dancing in a forest between giant trees during a rainy  weather."

# Generate an image with the given prompt
with torch.no_grad():
    image = pipe(prompt).images[0]

# Display the generated image
plt.imshow(image)
plt.axis('off')
plt.show()
```
